```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
games_table = pd.read_csv('games.csv')
print(games_table.head())
```

```python
games_table.columns = games_table.columns.str.lower()
print(games_table.head())
```

```python
# Convertendo colunas de texto existentes
games_table['name'] = games_table['name'].astype(str)
games_table['platform'] = games_table['platform'].astype(str)
games_table['genre'] = games_table['genre'].astype(str)
games_table['rating'] = games_table['rating'].astype(str)  # rating também é texto

# Convertendo ano
games_table['year_of_release'] = games_table['year_of_release'].fillna(0).astype(int)

# Convertendo vendas
games_table['na_sales'] = games_table['na_sales'].fillna(0).astype(float)
games_table['eu_sales'] = games_table['eu_sales'].fillna(0).astype(float)
games_table['jp_sales'] = games_table['jp_sales'].fillna(0).astype(float)
games_table['other_sales'] = games_table['other_sales'].fillna(0).astype(float)

# Convertendo notas
games_table['critic_score'] = pd.to_numeric(games_table['critic_score'], errors='coerce')
games_table['critic_score'] = games_table['critic_score'].fillna(games_table['critic_score'].mean())

games_table['user_score'] = pd.to_numeric(games_table['user_score'], errors='coerce')
games_table['user_score'] = games_table['user_score'].fillna(games_table['user_score'].mean())

# Conferindo tipos e primeiras linhas
print(games_table.dtypes)
print(games_table.head())
```

| Coluna                                            | Tipo original   | Novo tipo | Motivo da alteração                                                                                        |
| ------------------------------------------------- | --------------- | --------- | ---------------------------------------------------------------------------------------------------------- |
| `name`                                            | objeto/string   | string    | Garantir que todos os valores sejam tratados como texto, útil para filtragem e agrupamentos.               |
| `platform`                                        | objeto/string   | string    | Mantido como texto para categorizar plataformas.                                                           |
| `genre`                                           | objeto/string   | string    | Mantido como texto para análise por gênero.                                                                |
| `rating`                                          | objeto/string   | string    | Texto, facilita agrupamento e análise da classificação etária.                                             |
| `year_of_release`                                 | float           | int       | Converter para inteiro facilita análise temporal e agregações por ano.                                     |
| `na_sales`, `eu_sales`, `jp_sales`, `other_sales` | float           | float     | Mantido como float, mas substituímos NaN por 0 para permitir cálculos de vendas totais sem erros.          |
| `critic_score`                                    | float ou objeto | float     | Convertido para float, NaN preenchido com média da coluna para manter análise estatística consistente.     |
| `user_score`                                      | float ou objeto | float     | Convertido para float, NaN preenchido com média, valores “TBD” foram convertidos em NaN e depois tratados. |


2️⃣ Como lidar com valores ausentes
a) Abordagem escolhida

Para vendas (na_sales, eu_sales, jp_sales, other_sales) → substituímos NaN por 0, pois ausência de registro indica que não houve venda registrada naquela região.

Para notas (critic_score, user_score) → substituímos NaN pela média da coluna, porque notas faltantes podem distorcer análises de desempenho.

Para TBD (to be determined) → convertidos para NaN com pd.to_numeric(..., errors='coerce'), depois tratados conforme acima.

b) Possíveis razões para valores ausentes

Jogo antigo pode não ter recebido nota de crítica ou usuário ainda.

Falta de registro de vendas em algumas regiões.

Erro de coleta de dados ou sites que não informaram todas as informações.

Notas “TBD” indicam que a pontuação ainda não estava disponível no momento da coleta.

Passo 2 — Escolha do nível de significância (α)

Geralmente usamos α = 0.05 (5%).

Isso significa: temos 5% de chance de rejeitar H₀ incorretamente (erro tipo I).

Escolhemos 5% porque é um padrão na estatística: equilíbrio entre rigor e chance de erro.
